```python
# @title allFunctions
import random
def create_board(n):
    b = [['' for c in range(31)] for r in range(n)]
    return b
def barline():
    print("".join(f'{"────":>5}' for _ in range(10)))

def barlinePy(max):
    print(" ────  ",end="")
    print("".join("─" for _ in range(max+2)))

def barlinePy2(max):
    print(" ──────  ",end="")
    print("".join("─" for _ in range(max+2)))

def barline_py_board():
    print("+" + "─" * 202 + "+")


def wel_board():
    barline()
    start = 30
    for row in range(6):
      if row % 2 == 0:
          for cols in range(5):
            if 5 < start < 10:
              print(f"|    {start}",end="    ")
            else:
              print(f"|   {start}",end="    ")
            if cols == 4:
                print("|",end="")
            start -= 1
      else:
          start -= 4
          for cols in range(5):
            if 0 < start < 6:
              print(f"|    {start}",end="    ")
            else:
              print(f"|   {start}",end="    ")
            if cols == 4:
                print("|",end="")
            start += 1
          start -= 6
      print()
      barline()

def randSAndR(n):
    sket = ["🐍","🚀"]
    for row in range(n):
      for col in range(2,30):
        if random.randint(1,100) < 31  and row == 0:
          board[row][col] = random.choice(sket)
        elif board[0][col] == "🐍" or board[0][col] == "🚀":
          board[row][col] = board[0][col]

def resetboard(n):
    for row in range(n):
        for col in range(31):
            if col == 0:
                board[row][col] = "P" + str(row + 1)
            else:
                board[row][col] = str(col)

def print_board(n):
    print()
    for row in range(n):
      barline_py_board()
      for col in range(31):
          print(f"|{board[row][col]:^5}",end="")
      print("",end="|")
      print()
      barline_py_board()

def roll_dice():
    return random.randint(1, 6)

def play_game(n,lp,max_len):
    random.shuffle(lp)
    barlinePy(max_len)
    for i,v in enumerate(lp):
      print(f"| P{i+1} || {v:^{max_len}} |")
      barlinePy(max_len)

def end_game(w,max_len):
    barlinePy2(max_len)
    for i,v in enumerate(w):
      print(f"| No.{i+1} || {v:^{max_len}} |")
      barlinePy2(max_len)

def inputPy():
    while True:
      nump = input(f"Player {len(lp)+1} (or '-' to finish):  ")
      if nump == "-":
          break
      lp.append(nump)
      pp.append(0)
      loss.append(0)
      winner.append("")
    return len(lp)

def diceAll(n,w):
      row = 0
      roll = ""
      while row < n :
        if pp[row] < 30 and cnt < n:
          roll = input(f"  \nP{row+1}[{lp[row]}]'s turn. Press Enter d to roll the dice... ")
          while roll != "d" and roll != "D":
            roll = input(" Press Enter d to roll the dice... ")
          roll = ""
          r = roll_dice()
          print(f"   rolled a {r}")
          if(pp[row] + r <= 30):
            print(f"   moved from {pp[row]} to {pp[row]+r}")
          pp[row] += r
          chk_w = chkPoSingle(row)
          if chk_w != "":
            winner[cnt-1] = chk_w
        row = row + 1

def chkPoSingle(n):
    global cnt
    while True:
        if pp[n] >= 30:
            board[n][30] = "P" + str((n) + 1)
            cnt += 1
            print(f"   has finished the game!")
            return lp[n]
        elif pp[n] < 0:
            board[n][0] = "P" + str((n) + 1)
            pp[n] = 0
            print(f"   went back to start.")
            return ""
        elif board[n][pp[n]] == "🐍":
            snk = random.randint(1,3)
            print(f"   was bitten by a snake -{snk} position.")
            print(f"   move from {pp[n]} to {pp[n]-snk} ")
            pp[n] = pp[n] - snk
            loss[n] = 1
        elif board[n][pp[n]] == "🚀":
            rkt = random.randint(1,3)
            print(f"   found a rocket +{rkt} position")
            if(pp[n]+rkt <= 30):
              print(f"   move from {pp[n]} to {pp[n]+rkt} ")
            pp[n] = pp[n] + rkt
        else:
            board[n][pp[n]] = "P" + str((n) + 1)
            return ""

def remove(n):
  for row in range(n):
      if board[row][0] == "P" + str(row+1) and loss[row] != 1:
          board[row][0] = "S"
          continue
      for col in range(31):
        if col != pp[row] and loss[row] == 1:
          if col == 0:
            board[row][col] = "S"
          elif col > 0 and col < 30 and board[row][col] != "🐍" and board[row][col] != "🚀":
            board[row][col] = str(col)
        elif board[row][col] == "P" + str(row+1) and board[row][col] != "🐍" and board[row][col] != "🚀" and loss[row] != 1 and col < 30:
            board[row][col] = str(col)
            break
      loss[row] = 0
```

```python
# @title ข้อความอธิบายภาพเริ่มต้น
rematch = "y"
print(" Welcome to Snakes and Ladders!\n")
print("⋆˙✮🐍✮⋆˙⸆⸉")
wel_board()
while rematch == "y":
  lp = []
  pp = []
  loss = []
  winner = []
  cnt = 0
  a = inputPy()
  if(a == 0):
    break
  while(a == 1):
    print("Too lonely!!")
    a = inputPy()
  N = len(lp)
  board = create_board(N)
  resetboard(N)
  randSAndR(N)
  max_len = max(len(name) for name in lp)
  print()
  play_game(N,lp,max_len)
  print_board(N)
  while cnt < N:
    diceAll(N,winner)
    remove(N)
    print_board(N)
  end_game(winner,max_len)
  rematch = input("Wanna play again? y/n ")
print("\n\nThank you for playing Snakes and Ladders!")
```
